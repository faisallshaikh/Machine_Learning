```python
import pandas as pd
import numpy as np
```

```python
cat_cols = "cat1 cat1 cat2 cat3 cat2 np.nan cat3 cat3 cat2".split()
```

```python
d = {
    "name" : ["Minato","Itachi","Gohan","Vegeta","Eren","Might_Guy","Levi","Goku","Mikasa"],
    "Anime":["Naruto","Naruto","DBZ",np.nan,"AOT","Naruto","DBZ",np.nan,"AOT"],
    "Levels":[20,15,50,48,30,np.nan,50,np.nan,25],
    "cat_cols":cat_cols
}

df = pd.DataFrame(d)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
ord_cat = ["cat1","cat2","cat3","np.nan"]
```

```python
cat_features = ["cat_cols"]
cat_tranformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant",fill_value="missing")),
     ("ordinal", OrdinalEncoder(categories=[ord_cat]))
]) 

processor = ColumnTransformer(
        transformers=[("cat",cat_tranformer,cat_features)]
,remainder="passthrough")

model = Pipeline(steps=[
    ("processor", processor)
])

# X = df["Anime"]
# dff = processor.fit_transform(df)
# dff
```

```python
model
```

```python
tran = model.fit_transform(df)
```

```python
dff1 = pd.DataFrame(tran)
dff1
```

```python
anime_features = [2]
anime_tranformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant",fill_value="missing")),
     ("ordinal", OneHotEncoder(handle_unknown="ignore"))
]) 

processor = ColumnTransformer(
        transformers=[("cat",anime_tranformer,anime_features)]
,remainder="passthrough")

model1 = Pipeline(steps=[
    ("processor", processor)
])

dff2 = model1.fit_transform(dff1)
dff2
```

```python
pd.DataFrame(dff2)
```

```python

cat_features_1 = ["Anime","cat_cols"]

cat_imputer1 = SimpleImputer(strategy="constant", fill_value="missing")

tran = ColumnTransformer([
    ("imputer", cat_imputer1,cat_features_1)
],remainder="passthrough")

df1 = tran.fit_transform(df)
df1
```

```python
df
```

```python
new_df = pd.DataFrame(df1,columns=df.columns)
new_df
```

```python
one_hot = OneHotEncoder()
categorical = ["Anime","cat_cols"]
ohe = ColumnTransformer([("onehot",
                         one_hot,
                         categorical)],
                         remainder="passthrough")

coded_x = ohe.fit_transform(new_df)
coded_x
```

```python
pd.DataFrame(coded_x)
```
